```python
import pandas as pd
df = pd.read_excel('sales_data_sample.xlsx', sheet_name="saledata")
df.head(5)
```

```python
import matplotlib.pyplot as plt


# Tính tổng số lượng sản phẩm theo kích thước
quantity_by_size = df.groupby('DEALSIZE')['QUANTITYORDERED'].sum()

# Tính phần trăm
percentage_by_size = (quantity_by_size / quantity_by_size.sum()) * 100

# Vẽ biểu đồ tròn
plt.figure(figsize=(8,8))
plt.pie(percentage_by_size, labels=percentage_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Phần trăm số lượng sản phẩm theo kích thước')
plt.show()
```
